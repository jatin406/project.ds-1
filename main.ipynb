```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data load and cleaning of data
df =pd.read_csv(r"C:\Users\hp\Downloads\sales.csv")
df
```

```python
df.head()
```

```python
# now we will check for null values in the data set
df.isnull().sum()
```

```python
if df.isnull().sum().sum() == 0 :
    print("No null values in the dataset")
else :
    df = df.dropna()
    print("Null values have been dropped from the dataset")
```

```python
# step 2 :- basic analysis  and metric calculation
```

```python
total_sales =df['Total_Sales'].sum()
total_price = df['Price'].sum()
```

```python
product_wise = df.groupby("Product")["Total_Sales"].sum()
```

```python
print(f"\n Total sales : ${total_sales}")
```

```python
print(f"\n Total price : ${total_price}")
```

```python
print("Product - wise Sales Summary : - ")
print( product_wise)
```

```python
# step 3 --- chart 1 :- bar chart (product-wise sales)
```

```python
plt.Figure(figsize=(10,4))
product_wise.plot(kind = "bar" , color = ['green' , 'skyblue'])
plt.xlabel("Price")
plt.ylabel("Sales ($)")
plt.title("Total Product wise sales ")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.savefig("product_sales_bar.png")
plt.close()
```

```python
# step 4 -- creating a pie chart.
```

```python
#  a chart of Month - Wise sales DIstribution /---
```

```python
plt.Figure(figsize=(8,8))
product_wise.plot(kind = 'pie' ,color = 'orange' , autopct= "%0.0f%%" ,startangle = 140 ,)
plt.title("Monthly Sales Distribution share " , fontsize = 12)
plt.tight_layout()
plt.savefig("Monthly_Sales_Distribution.png")
```
